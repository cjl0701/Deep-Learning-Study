```python
# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러온다.
from keras.models import Sequential
from keras.layers import Dense

# 필요한 라이브러리를 디스크에서 메모리로 올린다.
import numpy
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분
seed=0
numpy.random.seed(seed)
tf.random.set_seed(seed)

# 준비된 수술 환자 데이터를 불러들임
Data_set = numpy.loadtxt("C:/Users\JAELYANG/dataset/ThoraricSurgery.csv", delimiter=",")

# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
X = Data_set[:,0:17]
Y = Data_set[:,17]

# 딥러닝 구조를 결정(모델을 설정하고 실행)
model = Sequential() # node layer를 일렬로
model.add(Dense(30,input_dim=17,activation='relu')) # 입력 데이터 노드 개수 17, 중간 층 노드 30
model.add(Dense(1,activation='sigmoid')) # 출력 노드 1개, 시그모이드 활성함수(0~1)

# 딥러닝 실행
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy']) # 모델을 컴퓨터가 알아듣게 컴파일
model.fit(X,Y,epochs=30, batch_size=10) # 트레이닝

# 결과 출력
print("\n Accuracy: %.4f" %(model.evaluate(X,Y)[1]))
```
